```python
from bs4 import BeautifulSoup
import requests
import csv
f = open("Midas_lab.text","w")



```

```python
content = requests.get('http://midas.iiitd.edu.in/').text
soup = BeautifulSoup(content,'lxml')
print(soup.prettify())
```

```python
for p in soup.find_all("p"):
    f.write("The despcription\r\n")
    f.write(p.text)
    print(p.text)
```

```python
f.write("The Website links")
for a in soup.find_all("img"):
    link = a['src']
    f.write(link)
    f.write("\n")
        
    print(link)
```
